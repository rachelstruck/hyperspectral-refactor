```python
illumination_data = pd.DataFrame({
    "wavelength" : [405, 488, 561, 637],
    "k"          : [1,   1,   1,   1]
})
illumination_wavelengths = np.array(illumination_data['wavelength'])
k = np.array(illumination_data['k'])
wavelength_range = (400,900)
bin_width = 10
fluorophore_list = get_fluorophore_list(wavelength_range=wavelength_range)
params = (illumination_wavelengths, k, wavelength_range, bin_width, fluorophore_list)
```

```python
bins_list = np.arange(*wavelength_range, bin_width)
bins_list
```

```python
N_ex = len(illumination_wavelengths)
N_em = len(bins_list)
idx = np.arange(N_ex*N_em)
```

```python
illumination_wavelengths = illumination_wavelengths[idx // N_em]
k = k[idx // N_em]
bins_list = bins_list[idx % N_em]
idx // N_em
```

```python
i = 0
fluorophore = fluorophore_list[i]
column_excitation = np.array(fluorophore.spectra.excitation[illumination_wavelengths])
fluorophore_emission_array = np.array(fluorophore.spectra.emission.loc[wavelength_range[0] : wavelength_range[1]-1])
column_emission = np.trapz(fluorophore_emission_array.reshape(N_em,bin_width))[idx % N_em]

k * fluorophore.brightness * column_excitation * column_emission
```
